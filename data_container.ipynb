列表的切片功能：arrp[start :stop :stop]
start为从第几个开始切，第二个stop为在哪里停止，第三个stop为步长，隔几个
计算机是从 0 开始数数的。

这三个参数都可以不写！Python 会根据情况自动补全。不写 start 默认从头开始，不写 stop 默认一直切到尾，不写 step 默认步长为 1（挨个切）。
Python 允许索引为负数，-1 代表倒数第一个元素，-2 代表倒数第二个。如果步长 (step) 是负数，意味着从右向左倒着切。

```python
nums =[1,2,3,4,5,6,7,8,9]
print(nums[2:4])
# 在4这个点截止不包括4,同理在某点开始则不包括那个点
print(nums[:4])
print(nums[4:])
print(nums[:])
print(nums[::2])
# 奇数
print(nums[1::2])
# 偶数
print(nums[-3:])
print(nums[::-1])
print(nums[-1::-1])
print(nums[-3::-1])
```

Python 中，lower 和 upper 是“方法”（Methods），也就是属于字符串的“动作工具”。

不加括号 ()：你只是指着这个工具告诉 Python：“看，这就是 lower 工具本身。” 所以 Python 返回了它的存储地址（也就是你看到的 <built-in method...>） 。

加上括号 ()：这是在告诉 Python：“请执行 lower 这个动作。” 只有加上括号，Python 才会去处理那个字符串。

对于元素容器还有高级的：元组 元组看作是**“绝对不可修改的列表”**。它的长相和列表很像，只是把方括号 [] 换成了圆括号 ()。

核心特性：一旦创建，里面的元素就不能增加、删除或修改（和字符串的不可变性一样）。

信工应用场景：当你需要传递一些绝对不能被篡改的配置信息时（比如网络通信中的 IP 地址和端口号组合），用元组最安全。'tuple' object does not support item assignment

字典 (Dictionary)：极速查找的键值对这是 Python 中最伟大、最常用的高级数据结构，没有之一。它用大括号 {} 包裹，里面存放的是成对的“键 (Key)”和“值 (Value)”。核心特性：它是无序的，但它的查找速度是极快的 $O(1)$ 级别。你不需要像列表那样从头到尾遍历去寻找数据，而是直接通过“名字（Key）”把数据一把抓出来

```python
interface_status ={
    "name":"eth0",
    "ip" : "10.5.5.0",
    "is_up":True 
}
print(interface_status["ip"])
interface_status["mac_address"] = "00:1A:2B"
print(interface_status["mac_address"])
```

集合 (Set)：自带数学光环的去重神器
集合也是用大括号 {} 包裹，但它里面只有值，没有键。它完美复刻了数学中“集合”的概念。

核心特性：天生去重，且无序。你往里面塞 100 个同样的数字，它最后也只会保留 1 个。

```python
visited_ips = ["1.1.1.1","2.2.2.2","1.1.1.1","3.3.3.3","2.2.2.2"]
unipue_ips = set(visited_ips)
print(unipue_ips) 
serve_a_ip = {"1.1.1.1","2.2.2.2"}
serve_b_ip = {"2.2.2.2","4.4.4.4."}
print(serve_a_ip & serve_b_ip)
```
